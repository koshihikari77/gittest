```python
def calc_variance(a):
    total = sum(a)
    length = len(a)
    mean = total/length
    variance = 0
    
    for i in a:
        variance += (i-mean)**2
    variance = variance/len(a)
    
    return variance**0.5
```

```python
b = [175,2,165,140,150]
b_variance = calc_variance(b)
print(b_variance)
```

```python

def long_repeat(line):
    """
        length the longest substring that consists of the same char
    """
    a = list(line)
    b = [-1]
    c = []
    
    for i in range(len(a)-1):
        if a[i] != a[i+1]:
            b.append(i)
    b.append(len(a)-1)
    
    for i in range(1,len(b)):
        c.append(b[i]-b[i-1])
    
    return max(c)
                 
long_repeat('aaaggikjlllll')
        
```

```python
def checkio(first, second):
    f = set(first.split(','))
    s = set(second.split(','))
    
    if len(f&s) == 0:
        return ""
    else:
        return ','.join(sorted(list(f&s)))

checkio("one,two,three", "four,five,one,two,six,three")
            
```
